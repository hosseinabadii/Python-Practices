```python
def fizz_buzz(*, fizz, buzz):
    answer = {}
    for num in range(1,101):
        result = num
        if num % fizz == 0:
            result = 'fizz'
        if num % buzz ==0:
            result = 'buzz'
        if num % fizz == 0 and num % buzz == 0:
            result = 'fizzbuzz'
        answer[num] = result    
    return answer
```

```python
output = fizz_buzz(fizz=4,buzz=7)
for i in output:
    print(f'{i:4d} : {output[i]}') 
```
