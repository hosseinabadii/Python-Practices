```python
for a in range(7801,8854):
    b = a // 1000
    bb = a % 1000

    c = bb // 100
    cc = bb % 100

    d = cc // 10
    dd = cc % 10
    sum1 = b + c + d + dd
    if sum1 % 13 == 0:
        print(f'sum of digits of {a} is {sum1}')
```

```python
for n in range(7801,8854):
    sum_ = 0
    num = n
    while n:
        sum_ += n % 10
        n = n // 10
    if sum_ % 13 == 0:
        print(f'sum of digits of {num} is {sum_}')
```
